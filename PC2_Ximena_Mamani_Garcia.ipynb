```python
HDILink1="https://docs.google.com/spreadsheets/d/e/2PACX-1vSp6si_gYJcZP5BMHVPp38FW0IaqH6uNkhHYVII2fKukg8jAMANm0tsDYzHyr0N_Q/pub?output=csv"

import pandas as pd
HDI=pd.read_csv(HDILink1)
HDI
```

```python
HDI.columns
```

```python
keep=['Back','Table 1. Human Development Index and its components ','Unnamed: 2','Unnamed: 4','Unnamed: 6','Unnamed: 8','Unnamed: 10','Unnamed: 12','Unnamed: 14']
dontKeep=set(HDI.columns.to_list())-set(keep)
HDICleaned=HDI.drop(columns=dontKeep)
HDICleaned
```

```python
HDICleaned.loc[:,keep]
```

```python
change={"Back":"HDI rank",'Table 1. Human Development Index and its components ':"Country","Unnamed: 2":"Human Development Index (HDI)","Unnamed: 4":"Life expectancy at birth","Unnamed: 6":"Expected years of schooling","Unnamed: 8":"Mean years of schooling","Unnamed: 10":"Gross national income (GNI) per capita","Unnamed: 12":"GNI per capita rank minus HDI rank	","Unnamed: 14":"HDI Rank"}
HDICleaned.rename(columns=change,inplace=True)
HDICleaned
```

```python
columns = pd.MultiIndex.from_tuples([
    ('HDI rank', '', ''),
    ('Country', '', ''),
    ('Human Development Index (HDI)', 'Value', '2021'),
    ('Life expectancy at birth', 'Years', '2021'),
    ('Expected years of schooling', 'Years', '2021'),
    ('Mean years of schooling', 'Years', '2021'),
    ('Gross national income (GNI) per capita', '2017 PPP $', '2021'),
    ('GNI per capita rank minus HDI rank', '', '2021'),
    ('HDI Rank', '', '2020')
])

# Para reemplazar las columnas del DataFrame actual
HDICleaned.columns = columns

HDICleaned
```

```python
#Se reestablece el índice del dataframe
HDICleaned = HDICleaned.reset_index(drop=True)

#Se procede a eliminar las filas por su posición
HDICleaned = HDICleaned.drop(index=[0,1,2,3,4,5,6])

HDICleaned
```

```python
HDICleaned = HDICleaned.drop('HDI rank', axis=1)
HDICleaned
```

```python
#Se reseteó el número de índice para que empieze desde el 1.
HDICleaned.reset_index(drop=True, inplace=True)
HDICleaned.index += 1

HDICleaned
```

```python
#Se escogió solo a la lista de países
HDICleaned = HDICleaned.iloc[:199]


HDICleaned
```

```python
keywords = ["HIGH HUMAN DEVELOPMENT", "OTHER COUNTRIES OR TERRITORIES", "LOW HUMAN DEVELOPMENT"]
HDICleaned = HDICleaned[~HDICleaned['Country'].isin(keywords)]

HDICleaned
```

```python
keep=[('Country', '', ''), ('Human Development Index (HDI)', 'Value', '2021'), ('Life expectancy at birth', 'Years', '2021'), ('Expected years of schooling', 'Years', '2021'), ('Mean years of schooling', 'Years', '2021'), ('Gross national income (GNI) per capita', '2017 PPP $', '2021'), ('GNI per capita rank minus HDI rank', '', '2021'), ('HDI Rank','','2020')]
HDICleaned=HDICleaned.loc[:,keep]
HDICleaned
```

```python
HDICleaned.info()
```

```python
objectHDI=HDICleaned.select_dtypes(include=['object']).columns
objectHDI
```

```python
#Se reemplaza la coma por el punto para tener listas numéricas
HDICleaned[('Human Development Index (HDI)','Value','2021')]=HDICleaned[('Human Development Index (HDI)','Value','2021')].str.replace(",",'.')
HDICleaned[('Life expectancy at birth','Years','2021')]=HDICleaned[('Life expectancy at birth','Years','2021')].str.replace(",",'.')
HDICleaned[('Expected years of schooling','Years','2021')]=HDICleaned[('Expected years of schooling','Years','2021')].str.replace(",",'.')
HDICleaned[('Mean years of schooling','Years','2021')]=HDICleaned[('Mean years of schooling','Years','2021')].str.replace(",",'.')
HDICleaned
```

```python
HDICleaned.info()
```

```python
HDICleaned[('Human Development Index (HDI)','Value','2021')] = pd.to_numeric(HDICleaned[('Human Development Index (HDI)','Value','2021')], errors='coerce')
HDICleaned[('Life expectancy at birth','Years','2021')] = pd.to_numeric(HDICleaned[('Life expectancy at birth','Years','2021')], errors='coerce')
HDICleaned[('Expected years of schooling','Years','2021')] = pd.to_numeric(HDICleaned[('Expected years of schooling','Years','2021')], errors='coerce')
HDICleaned[('Mean years of schooling','Years','2021')] = pd.to_numeric(HDICleaned[('Mean years of schooling','Years','2021')], errors='coerce')
HDICleaned[('Gross national income (GNI) per capita','2017 PPP $','2021')] = pd.to_numeric(HDICleaned[('Gross national income (GNI) per capita','2017 PPP $','2021')], errors='coerce')
HDICleaned[('GNI per capita rank minus HDI rank','','2021')] = pd.to_numeric(HDICleaned[('GNI per capita rank minus HDI rank','','2021')], errors='coerce')
HDICleaned[('HDI Rank','','2020')] = pd.to_numeric(HDICleaned[('HDI Rank','','2020')], errors='coerce')
# Para verificar los tipos de datos
print(HDICleaned.dtypes)
```

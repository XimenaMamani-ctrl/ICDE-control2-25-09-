```python
Lista="https://docs.google.com/spreadsheets/d/e/2PACX-1vS1uz3AbIpCrcMHIPfQ0GxCTHX9v4kkA3EiCzQp7GBsPVYQFsUm7WnjhzqFD1WFgjO5es1s2i0EyapJ/pub?output=csv"

import pandas as pd
FacellidosCOVID=pd.read_csv(Lista)
FacellidosCOVID
```

```python
FacellidosCOVID.info()
```

```python
FacellidosCOVID['FECHA_FALLECIMIENTO'] = pd.to_datetime(FacellidosCOVID['FECHA_FALLECIMIENTO'], format='%Y%m%d', errors='coerce')

#Se craron nuevas tablas de año y mes
FacellidosCOVID['AÑO'] = FacellidosCOVID['FECHA_FALLECIMIENTO'].dt.year
FacellidosCOVID['MES'] = FacellidosCOVID['FECHA_FALLECIMIENTO'].dt.month

#Se agrupan por año y mes y además se coeuntan los fallecidos
reporte_mensual = FacellidosCOVID.groupby(['AÑO', 'MES']).size().reset_index(name='TOTAL_FALLECIDOS')

#wide to long usando year and month como fila
reporte_mensual['MES_AÑO'] = reporte_mensual['AÑO'].astype(str) + '-' + reporte_mensual['MES'].astype(str).str.zfill(2)

# Seleccionar solo la columna "MES_AÑO" y "TOTAL_FALLECIDOS"
reporte_final = reporte_mensual[['MES_AÑO', 'TOTAL_FALLECIDOS']]

reporte_final
```

```python
#Long to wide

# Convertir la columna 'FECHA_FALLECIMIENTO' a formato datetime
FacellidosCOVID['FECHA_FALLECIMIENTO'] = pd.to_datetime(FacellidosCOVID['FECHA_FALLECIMIENTO'], format='%Y%m%d', errors='coerce')

# Crear un reporte de fallecidos por mes con columnas separadas para cada tipo de clasificación
reporte_mensual_clasificacion = FacellidosCOVID.pivot_table(
    index=FacellidosCOVID['FECHA_FALLECIMIENTO'].dt.to_period('M'),  # Agrupar por mes
    columns='CLASIFICACION_DEF',  # Crear columnas separadas para cada clasificación de defución
    aggfunc='size',  # Contar la cantidad de fallecidos por cada clasificación
    fill_value=0  # Rellenar con 0 donde no haya datos en vez
).reset_index()

# Mostrar el resultado
reporte_mensual_clasificacion
```

```python
# Extraer solo el mes de la columna 'FECHA_FALLECIMIENTO'
FacellidosCOVID['MES'] = FacellidosCOVID['FECHA_FALLECIMIENTO'].dt.month

# Crear un reporte de fallecidos por mes con columnas separadas para cada tipo de clasificación
reporte_mensual_clasificacion = FacellidosCOVID.pivot_table(
    index='MES',  # Agrupar por mes
    columns='CLASIFICACION_DEF',  # Crear columnas separadas para cada clasificación de defunción
    aggfunc='size',  # Contar la cantidad de fallecidos por cada clasificación
    fill_value=0  # Rellenar con 0 donde no haya datos
).reset_index()

# Mostrar el resultado
reporte_mensual_clasificacion
```
